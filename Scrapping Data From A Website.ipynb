# **Scraping Data From a Table on a Website**



```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://webpages.charlotte.edu/mscipion/"

r= requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
table = soup.find("table", class_="table table-sm table-hover screenertable")
print(table)     #uncomment to see full outp
```

```python

if "soup" not in locals() or soup is None:
    print("Soup object is not defined or is None.")
else:
    # Find the table within the soup
    table = soup.find("table")
    if table is not None:
        headers = table.find_all("th")
        print(headers)
    else:
        print("No table found in the soup object.")
```

```python
titles = []

for i in headers:
    title = i.text
    titles.append(title)
    
print(titles)
```

```python
df= pd.DataFrame(columns=titles)
print(df)
```

```python
rows = table.find_all("tr")
print(rows)           #uncomment to see full output
```

```python

for i in rows[1:]:
    print(i) 
```

```python

for i in rows[1:]:
    data = table.find_all("td")
    print(data)
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text.strip() for tr in data]
    df.loc[len(df)] = row

df.head(6)
```
